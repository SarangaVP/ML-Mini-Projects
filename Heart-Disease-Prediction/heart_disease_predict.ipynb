```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_disease_data = pd.read_csv('heart_disease_data.csv')
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x='target', data=heart_disease_data)
plt.title('Heart disease(1) or not(0)')
plt.show()
```

Split features and target

```python
X = heart_disease_data.drop(columns='target', axis=1)
Y = heart_disease_data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

Train the model

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_pred, Y_test)
print(f"Accuracy: {accuracy}")
```

```python
new_data = [66, 1, 0, 138, 191, 1, 1, 145, 0, 3.2, 1, 2, 1]
data_np = np.asarray(new_data).reshape(1, -1)

prediction = model.predict(data_np)[0]
if prediction == 1:
    print('Heart Disease')
else:
    print('No Heart Disease')
```
